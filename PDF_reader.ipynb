```python
import pyttsx3                   # For text-to-speech conversion
from PyPDF2 import PdfReader     # For reading and extracting text from PDF files

# Ask user to provide the PDF file path (local or URL)
file = input("Enter Path:")

# Ask user to provide the page number they want to read
# Note: page numbers start from 0 (first page = 0)
i = int(input("Enter Page Number:"))
reader = PdfReader(file)   # Load the PDF file
pages = reader.pages                    
print(pages)

melo = pyttsx3.init()     # Initialize the text-to-speech engine
page = reader.pages[i]    # Extract text from the selected page
text = page.extract_text()             
melo.say(text)           # Convert the extracted text into speech
melo.runAndWait()        # Run the speech engine to actually speak out loud
```
